## Quais são as 5 principais cidades da realização da pesquisa?

```python
plt.figure(figsize=(15,8))

df['City'].value_counts().iloc[:5].plot.pie(autopct="%1.2f%%",fontsize=13,startangle=90,labels=['']*5,
                                         cmap='Set2',explode=[0.05] * 5,pctdistance=1.2)
plt.ylabel("")
plt.legend(loc='upper left',labels=df['City'].value_counts().iloc[:5].index)
plt.show()
```

## Gráfico dos gêneros que atuam na área de TI

```python
plt.figure(figsize=(15,8))

df['Gender'].value_counts().plot.pie(autopct="%1.2f%%",fontsize=12,startangle=90,
                                         cmap='crest',explode=[0.05] * 3,pctdistance=1.1,
                                          labeldistance=1.3,textprops={'fontsize': 15})
plt.ylabel("")
plt.show()
```

## Linguagens de Programação mais utilizadas

```python
modi_df = df.copy()
modi_df['Your main technology / programming language'] = modi_df['Your main technology / programming language'].str.lower()
modi_df['Your main technology / programming language'] = modi_df['Your main technology / programming language'].str.strip()

top_tech_sal = modi_df['Yearly brutto salary (without bonus and stocks) in EUR'].groupby(
            modi_df['Your main technology / programming language']).sum().sort_values(ascending=False).index.to_list()[0:15]
top_tech_sal
```
